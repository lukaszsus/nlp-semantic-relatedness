```python
import pandas as pd
import os
import sys
import inspect
import seaborn as sns
import matplotlib.pyplot as plt
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, parentdir)
from settings import PROJECT_PATH
```

## Load results

```python
embeddings_results_path = "output/embeddings_results.csv"
wordnet_results_path = "output/wordnet_results_20191118.csv"
embeddings_results_path = os.path.join(PROJECT_PATH, embeddings_results_path)
wordnet_results_path = os.path.join(PROJECT_PATH, wordnet_results_path)

model_names_mapper = {"w2v-nkjp+wiki-forms-restricted-100-cbow-hs.txt-filtered": "w2v-100-cbow",
                     "w2v-nkjp+wiki-forms-restricted-100-skipg-hs.txt-filtered": "w2v-100-skipg",
                     "w2v-nkjp+wiki-forms-restricted-300-cbow-hs.txt-filtered": "w2v-300-cbow",
                     "w2v-nkjp+wiki-forms-restricted-300-skipg-hs.txt-filtered": "w2v-300-skipg",
                     "fasttext-kgr10.plain.lemma.lower.cbow.dim100.neg10.bin-filtered": "fasttext-100-cbow",
                     "slowosiec-graph-hiperonim-connected-by-top-sort": "slowosiec"}
```

```python
df_embeddings = pd.read_csv(embeddings_results_path, index_col=0)
df_embeddings.sort_values(by="similarity_cor", ascending=False)
```

```python
df_wordnet = pd.read_csv(wordnet_results_path, index_col=0)
df_wordnet.sort_values(by="similarity_cor", ascending=False)
```

# Word 2 Vec

## Similarity measures comparison

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

df = df_embeddings[["metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["metric", "similarity_func"]).mean()
df = df.reset_index()
sns.barplot(ax=axes[0, 0], x="similarity_func", y="similarity_cor", hue="metric", data=df)
axes[0, 0].set_title("Mean similarity correlation per similarity function")

df = df_embeddings[["metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["metric", "similarity_func"]).mean()
df = df.reset_index()
sns.barplot(ax=axes[1, 0], x="similarity_func", y="relatedness_cor", hue="metric", data=df)
axes[1, 0].set_title("Mean relatedness correlation per similarity function")

df = df_embeddings[["metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["metric", "similarity_func"]).max()
df = df.reset_index()
sns.barplot(ax=axes[0, 1], x="similarity_func", y="similarity_cor", hue="metric", data=df)
axes[0, 1].set_title("Max similarity correlation per similarity function")

df = df_embeddings[["metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["metric", "similarity_func"]).max()
df = df.reset_index()
sns.barplot(ax=axes[1, 1], x="similarity_func", y="relatedness_cor", hue="metric", data=df)
axes[1, 1].set_title("Max relatedness correlation per similarity function")
```

## Model comparison

```python
def plot_model_comp(similarity_func):
    df_func = df_embeddings[df_embeddings["similarity_func"] == similarity_func]
    df_func.loc[:, "model_name"] = df_func["model_name"].map(model_names_mapper)

    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    df = df_func[["model_name", "metric", "similarity_func", "similarity_cor", "relatedness_cor"]]
    df = df.reset_index()
    df = df.sort_values(by="model_name")
    sns.barplot(ax=axes[0], x="model_name", y="similarity_cor", hue="metric", data=df)
    axes[0].set_title("Similarity correlation per model using {}".format(similarity_func))
    axes[0].set_xticklabels(
        axes[0].get_xticklabels(), 
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='x-large'
    )
    
    df = df_func[["model_name", "metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["model_name", "metric", "similarity_func"]).max()
    df = df.reset_index()
    df = df.sort_values(by="model_name")
    sns.barplot(ax=axes[1], x="model_name", y="relatedness_cor", hue="metric", data=df)
    axes[1].set_title("Relatedness correlation per model using {}".format(similarity_func))
    axes[1].set_xticklabels(
        axes[1].get_xticklabels(), 
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='x-large'
    )
```

```python
plot_model_comp("cosine")
```

```python
plot_model_comp("euclidean")
```

```python
plot_model_comp("manhattan")
```

# Relatedness in WordNet

```python
wordnet_research_path = "output/wordnet_results.csv"
wordnet_research_path = os.path.join(PROJECT_PATH, wordnet_results_path)
wordnet_model_names_mapper = {"slowosiec-graph-hiperonim-connected-by-top-sort": "hiperonim",
                     "slowosiec-graph-lexical-connected-by-top-sort": "lexical",
                     "slowosiec-graph-hiperonim-lexical-connected-by-top-sort": "hiperonim+lexical",
                     "slowosiec-graph-hiperonim-lexical-selected-connected-by-top-sort": "hiperonim+lexical-selected"}
```

```python
df_wordnet_research = pd.read_csv(wordnet_research_path, index_col=0)
```

```python
df_wordnet_research.loc[:, "model_name"] = df_wordnet_research["model_name"].map(wordnet_model_names_mapper)
```

```python
def compare_wordnets(df_wordnets, similarity_func):
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    df_wordnets = df_wordnets[df_wordnets["similarity_func"]==similarity_func]
    df = df_wordnets[["model_name", "metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["metric", "model_name"]).max()
    df = df.reset_index()
    df = df.sort_values(by="model_name")
    sns.barplot(ax=axes[0], x="model_name", y="similarity_cor", hue="metric", data=df)
    axes[0].set_title("Similarity correlation per model using {}".format(similarity_func))
    axes[0].set_xticklabels(
        axes[0].get_xticklabels(), 
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='x-large'
    )
    
    df = df_wordnets[["model_name", "metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["metric", "model_name"]).max()
    df = df.reset_index()
    df = df.sort_values(by="model_name")
    sns.barplot(ax=axes[1], x="model_name", y="relatedness_cor", hue="metric", data=df)
    axes[1].set_title("Relatedness correlation per model using {}".format(similarity_func))
    axes[1].set_xticklabels(
        axes[1].get_xticklabels(), 
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='x-large'
    )
```

```python
df_wordnet_research.sort_values(by="relatedness_cor", ascending=False)
```

```python
compare_wordnets(df_wordnet_research, "LeacockChodorow")
```

```python
compare_wordnets(df_wordnet_research, "WuPalmer")
```
