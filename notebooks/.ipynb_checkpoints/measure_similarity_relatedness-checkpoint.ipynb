## Similarity measures comparison

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

df = df_embeddings[["metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["metric", "similarity_func"]).mean()
df = df.reset_index()
sns.barplot(ax=axes[0, 0], x="similarity_func", y="similarity_cor", hue="metric", data=df)
axes[0, 0].set_title("Mean similarity correlation per similarity function")

df = df_embeddings[["metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["metric", "similarity_func"]).mean()
df = df.reset_index()
sns.barplot(ax=axes[1, 0], x="similarity_func", y="relatedness_cor", hue="metric", data=df)
axes[1, 0].set_title("Mean relatedness correlation per similarity function")

df = df_embeddings[["metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["metric", "similarity_func"]).max()
df = df.reset_index()
sns.barplot(ax=axes[0, 1], x="similarity_func", y="similarity_cor", hue="metric", data=df)
axes[0, 1].set_title("Max similarity correlation per similarity function")

df = df_embeddings[["metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["metric", "similarity_func"]).max()
df = df.reset_index()
sns.barplot(ax=axes[1, 1], x="similarity_func", y="relatedness_cor", hue="metric", data=df)
axes[1, 1].set_title("Max relatedness correlation per similarity function")
```

## Model comparison

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

df_func = df_embeddings[df_embeddings["similarity_func"] == 'cosine']

df = df_func[["model_name", "metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["model_name", "metric", "similarity_func"]).mean()
df = df.reset_index()
sns.barplot(ax=axes[0, 0], x="model_name", y="similarity_cor", hue="metric", data=df)
axes[0, 0].set_title("Mean similarity correlation per model using cosine")
axes[0, 0].set_xticklabels([])

df = df_func[["model_name", "metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["model_name", "metric", "similarity_func"]).mean()
df = df.reset_index()
sns.barplot(ax=axes[1, 0], x="model_name", y="relatedness_cor", hue="metric", data=df)
axes[1, 0].set_title("Mean relatedness correlation per model using cosine")
axes[1, 0].set_xticklabels(
    axes[1, 0].get_xticklabels(), 
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'
)

df = df_func[["model_name", "metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["model_name", "metric", "similarity_func"]).max()
df = df.reset_index()
sns.barplot(ax=axes[0, 1], x="model_name", y="similarity_cor", hue="metric", data=df)
axes[0, 1].set_title("Max similarity correlation per model using cosine")
axes[0, 1].set_xticklabels([])

df = df_func[["model_name", "metric", "similarity_func", "similarity_cor", "relatedness_cor"]].groupby(["model_name", "metric", "similarity_func"]).max()
df = df.reset_index()
sns.barplot(ax=axes[1, 1], x="model_name", y="relatedness_cor", hue="metric", data=df)
axes[1, 1].set_title("Max relatedness correlation per model using cosine")
axes[1, 1].set_xticklabels(
    axes[1, 1].get_xticklabels(), 
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'
)
```

```python
def predict_encoding(file_path, n_lines=20):
    '''Predict a file's encoding using chardet'''
    import chardet

    # Open the file as binary data
    with open(file_path, 'rb') as f:
        # Join binary lines for specified number of lines
        rawdata = b''.join([f.readline() for _ in range(n_lines)])

    return chardet.detect(rawdata)['encoding']
```
